```python
## Creamos una copia del DF original
df_hurtos_MDE_Limp = df_hurtos_MDE
```

```python
## Reemplazamos y guardamos los cambios en el mismo DF_limpio, esto lo indicamos con el
## inplace = True
df_hurtos_MDE_Limp.nombre_barrio.replace(
     { "U.D. Atanasio Girardot":"Estadio",
       "La Palma S.C.":"La Palma",
       "Naranjal S.C.":"Naranjal",
       "San Pablo A.V.":"San Pablo",
       "Las Palmas S.E.":"Las Palmas",
       "Las Playas S.C.":"Las Playas",
       "La Florida S.A.P.":"La Florida",
       "La Esperanza A.V.":"La Esperanza",
       "Facultad de Minas U. Nal":"Facultad de Minas U. Nacional",
       "Pajarito S.C.":"Pajarito",
       "Cerro Nutibara Ins.":"Cerro Nutibara",
       "Batallón Cuarta Brigada":"Cuarta Brigada"}, inplace=True)
```

```python
type(df_hurtos_MDE_Limp)   # Verifico que la copia si sea un dataframe
```

```python
## Valido si efectivamente se hicieron los cambios
pd.unique(df_hurtos_MDE_Limp.nombre_barrio.sort_values())
```

```python
df_hurtos_MDE_Limp.info()
```

```python
## Cambiar tipo de dato de algunos campos
df_hurtos_MDE_Limp = df_hurtos_MDE_Limp.astype({'edad': 'uint8',
                                                'cantidad':'uint8',
                                                'estado_civil':'category',
                                                'sexo':'category',
                                                'bien':'category',
                                                'actividad_delictiva':'category',
                                                'modelo':'category',
                                                'sede_receptora':'category',
                                                'color':'category',
                                                'modalidad':'category'
                           })
df_hurtos_MDE_Limp.info()  ## Valido los cambios realizados
```

```python
## Agrupamos por sexo, y almacenamos el resultado en una variable
df_hurtos_MDE_sexo = df_hurtos_MDE_Limp.groupby("sexo").size()
df_hurtos_MDE_sexo
```

```python
## Creamos el tamaño de la grafica
plt.figure(figsize=(10,6))
## Creamos histograma con hue para clasificar las edades por sexo
## el parametro multiple, me permite apilar los valores
sns.histplot(data=df_hurtos_MDE_Limp, x='edad', hue='sexo', multiple = "stack")
## Establezco un titulo para el grafico
plt.title("Distribución de edades por sexo de las victimas")
## Establezco titulo para el eje X
plt.xlabel("Edad")
## Establezco titulo para el eje Y
plt.ylabel("Numero de victimas")
## Creamos malla
plt.grid()
## Imprima grafica en pantalla
plt.show()
```

```python
## GRAFICO DE CAJAS, permite visualizar la media, la mediana, los percentiles...
## ... y los outliers (Valores atipicos)
## Creamos el tamaño de la grafica
plt.figure(figsize=(10,6))
## Creamos cajas con variables categoricas como sexo para analizar por edades
sns.boxplot(data=df_hurtos_MDE_Limp,  x='sexo', y='edad')
## Establezco un titulo para el grafico
plt.title("Analisis de sexo por edades de las victimas")
## Establezco titulo para el eje X
plt.xlabel("Sexo")
## Establezco titulo para el eje Y
plt.ylabel("Edad")
## Imprima grafica en pantalla
plt.show()
```

```python
## GRAFICO DE CAJAS, permite visualizar la media, la mediana, los percentiles...
## ... y los outliers (Valores atipicos)
## Creamos el tamaño de la grafica
plt.figure(figsize=(10,6))
## Creamos cajas con variables categoricas como sexo para analizar por edades
sns.boxplot(data=df_hurtos_MDE_Limp,  x='modalidad', y='edad')
## Establezco un titulo para el grafico
plt.title("Analisis de modalidad por edades de las victimas")
## Establezco titulo para el eje X
plt.xlabel("Modalidad")
plt.xticks(rotation=90)  ## Rotacion de eje X
## Establezco titulo para el eje Y
plt.ylabel("Edad")
## Imprima grafica en pantalla
plt.show()
```
