**Introduccion a Numpy**

En un Numpy existen estructuras de datos llamadas arreglos , donde son aquellos que tienen 1 un indice y tiene varias dimensiones entre esas son:

Arreglo Unidimensional: Tiene una dimension

Arreglo Bidimensional: Tiene 2 dimensiones

Arreglo Tridimensional: Tiene 3 dimensiones

```python
# Importamos la libreria de numpy
import numpy as np

# Dimensionamos el arreglo 1 dimension
vector = np.array([1,4,5,7,8,9,4,3,8,9,4])
print(vector)

# Imprimir la longitud del arreglo
print(len(vector))

#Indexacion del vector
print(vector[3])

#Actualizar el valor del vector o arreglo unidimensional
vector[3] =50
print(vector)

# Slicing es recorrer el indice del arreglo [indice_inicial: indice_final-1:paso]
print(vector[0:10])
print(vector[0:10:2])

# Ciclos
for valor in vector:
  print(valor)



```

```python
#Atributos

# Atributos del arreglo
print('Tipo de Arreglo',vector.dtype)
print('Tamano del arreglo', vector.size)
print('Dimensiones del arreglo',vector.ndim)
print('Elemento de cada dimension', vector.shape)

```

```python
#Metodos

# Definimos la lista
lista = [1,2,3,4,5]
vector= np.array(lista)

print(vector, type(vector))

# Multiplicacion entre arreglos
arreglouno = np.array([2,4,6,8])
arreglodos = np.array([3,5,7,9])
print(arreglouno*arreglodos)

```

```python
# Metodos estadisticos
print("Suma es: ", vector.sum())
print("Promedio es:", vector.mean())
print("Valor maximo: ", vector.max())
print("Valor minimo: ", vector.min())
```

```python
# Arreglos de 2 dimensiones
matriz = np.array(([1,2,5,4],[9,5,5,7],[6,0,0,8],[3,9,6,5]))
print(matriz)

# Cantidad de elementos de la matriz
print(len(matriz))

# Indexacion de la matriz
print(matriz[2,2])

# Actualizar un valor en la matriz
matriz[2,2] =7

print(matriz)
```

```python
# Atributos de arreglos 2D
print('Tipo de matriz: ', matriz.dtype)
print('Tamano de la matriz', matriz.size)
print('Dimensiones del arreglo', matriz.ndim)
print('Elementos de cada dimension de la matriz: ', matriz.shape)
```

```python
# Metodo de np.array() - crear una a partir de una seucencia de datos
lista= [[1,2,5,4],[9,5,5,7], [6,0,0,8],[3,9,6,5]]
matriz =(np.array(lista))
print(matriz, type(matriz))

# Metodos estadisticos
print('Suma es: ',matriz.sum())
print('Promedio es: ', matriz.mean())
print('Valor maximo', matriz.max())
print("Valor minimo", matriz.min())

#  Metodos propios 2D
# Cambiar la forma de la matriz de 4x4 a 2x8
matriz_reshape=matriz.reshape(2,8)
print(" \nMatriz con forma cambiada:")
print(matriz_reshape)

# Transponer matriz
print("\nMatriz transpuesta:")
matriz_transpose= matriz_reshape.transpose()
print(matriz_transpose)
```

```python
vectorUno =np.array([1,2,3])
vectorDos= np.array([4,5,6])

#Suma de vectores
suma=vectorUno+vectorDos
print(suma, "\n")

sumaDos=vectorUno+6
print(sumaDos,"\n")

# Producto punto
producto= np.dot(vectorUno, vectorDos)
print(producto,"\n")
```

```python
import pandas as pd
#Cargar datos desde github
url = 'https://github.com/carlosgomez0717/Machine-Learning/raw/main/Calificaciones.xlsx'

datasetExcelURL = pd.read_excel(url, sheet_name='Hoja1')

datasetExcelURL

# Muestra los primeros 5 registros del dataset
datasetExcelURL.head()

# Muestras los ultimos 5 registros del dataset
datasetExcelURL.tail()

# Muestra informacion tecnica de las columnas del dataset
datasetExcelURL.info()

# Muestra informacion estadisitica del dataset
datasetExcelURL.describe()
```

```python
# Transformacion de los datos
# Limpieza de datos: Es el proceso de transformacion de los datos con el fin de obtener una
# mejor calidad de la informacion para obtener un excelente analisis.

# Rellenar las filas con valores nulos en un columna en especifica
datasetNuevo = datasetExcelURL['Idiomas']
datasetNuevo= datasetNuevo.fillna(0.0)
datasetNuevo
```

```python
# Eliminar las filas con valores nulos del dataset completo
datasetNuevo =datasetExcelURL.dropna()
datasetNuevo
```

```python
# Libreria para importar graficas
import matplotlib.pyplot as plt

# Crear un grafico a a partir del dataset
datasetNuevo.plot()
plt.show()
```

```python
# Agrupar datos

# 1 Cantidad de estudiantes por programa
est_programa= datasetNuevo.groupby('Programa').count()
est_programa
est_programa=datasetNuevo.groupby('Programa').agg({'Programa':'count'})
est_programa
est_programa = datasetNuevo.groupby('Programa') # Opc03
print(est_programa.size(),'\n')
```

```python
# Cantidad estudiantes por genero
est_genero=datasetNuevo.groupby('Genero').agg({'Genero':'count'})
est_genero
```

```python
# Cantidad de estudiante por programa y genero
est_programa_genero=datasetNuevo.groupby(['Programa','Genero']).agg({'Programa':'count'})
est_programa_genero
```

```python
# Cantidad de estudiantes que aprueban matematicas
aprueba_matematicas=datasetNuevo['Matematicas']>=4.0
aprueba_matematicas=aprueba_matematicas.replace({False:'No aprobado', True:'Aprobado'})
estado_aprueba_matematicas = datasetNuevo.groupby(aprueba_matematicas)['Matematicas'].count()
estado_aprueba_matematicas
```

```python
# Operaciones y generacion de nuevas columnas
datasetNuevo['Promedio']= (datasetNuevo['Matematicas']+datasetNuevo['Español']+datasetNuevo['Ciencias']+datasetNuevo['Idiomas'])/4
datasetNuevo
```

```python
# 06. Tablas de resumen o tablas dinamicas
tabla_dinamica = datasetNuevo.pivot_table(
    values=['Matematicas', 'Español', 'Ciencias', 'Idiomas'],
    index=['Programa', 'Genero'],
    aggfunc='mean',
)
tabla_dinamica
```
